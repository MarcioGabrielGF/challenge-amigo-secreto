# 📊 Análise das Lojas - Alura Store
Durante este desafio, vamos ajudar o Senhor João a decidir qual loja vender.

### Objetivo
- Analisar os dados de vendas, faturamento, categorias mais vendidas, avaliações e frete.
- Criar gráficos para visualizar os resultados.
- Recomendar qual loja possui menor eficiência.

## 1. Carregando os dados

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar dados
df = pd.read_csv('alurastore_dados.csv')

df.head()
```

## 2. Faturamento por Loja

```python
import seaborn as sns

faturamento_loja = df.groupby('Loja')['Faturamento'].sum().sort_values()

plt.figure(figsize=(7,7))
plt.pie(faturamento_loja, labels=faturamento_loja.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))
plt.title('Faturamento Total por Loja', fontsize=16)
plt.show()

faturamento_loja
```

## 3. Categorias mais vendidas por Loja

```python
categorias_loja = df.groupby(['Loja','Categoria'])['Quantidade_Vendida'].sum().unstack()
categorias_loja.plot(kind='bar', figsize=(10,6), stacked=True)
plt.title('Categorias mais vendidas por Loja')
plt.ylabel('Quantidade Vendida')
plt.show()

categorias_loja
```

## 4. Avaliação média por Loja

```python
avaliacao_media = df.groupby('Loja')['Avaliacao'].mean().sort_values()
avaliacao_media.plot(kind='bar', figsize=(8,5), color='orange', title='Avaliação Média por Loja')
plt.ylabel('Nota (1 a 5)')
plt.show()

avaliacao_media
```

## 5. Frete médio por Loja

```python
frete_medio = df.groupby('Loja')['Frete'].mean().sort_values()
frete_medio.plot(kind='line', figsize=(8,5), color='green', title='Frete Médio por Loja')
plt.ylabel('Frete Médio (R$)')
plt.show()

frete_medio
```
